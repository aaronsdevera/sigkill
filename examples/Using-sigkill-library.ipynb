# Using `sigkill` library

Import the `sigkill` library.

```python
from sigkill import SigKill
```

Create a `Sigkill` object.

```python
sk = SigKill()
```

Write an aggregation in SQL, execute it, and print the resulting Dataframe..

```python
SQL = '''select
    conversationId,
    count(*) volume
    from messages
    group by 1
    order by volume desc
'''
df = sk.db().table('messages').get_sql(SQL, ['conversationId','volume']).to_df()
df.head()
```

Plot it!

```python
df.head(10).plot.bar(x='conversationId', y='volume', rot=0, figsize=(20,10))
```
